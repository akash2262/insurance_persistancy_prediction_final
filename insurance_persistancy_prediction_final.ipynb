Shapiro-Wilk test is a statistical test used to assess whether a given sample comes from a normally distributed population.

```python
from scipy.stats import shapiro

_, p_value = shapiro(df['Mapped Agent 13M Persistency'])
print("Shapiro-Wilk p-value:", p_value)

if p_value < 0.05:
    print("The data does not follow a normal distribution.")
else:
    print("The data may follow a normal distribution.")
```

Defining a custom mapping of education categories using a dictionary (custom_categories). It then applies this mapping to a column named 'Application Life Assured Education' in a DataFrame to create a new column named 'Custom_Education_Category

```python
custom_categories ={
    'UG': ['Graduation', 'S S C', 'B A', 'B E', 'B Sc', 'B Com', 'B Ed', 'BCA', 'B B A', 'BDS', 'BHMS', 'Bachelor of Veterinary Science'],
    'Medical': ['M Pharm', 'BAMS', 'BHMS', 'M D', 'MBBS', 'Diploma in Pharmacy', 'B Pharm', 'Diploma in General Nursing'],
    'Others': ['Others','Other', 'BAC', 'MPY', 'PharmD', 'DMLT', 'DTE', 'PBM','BUMS'],
    'Engineering': ['B Tech', 'M Tech', 'DCE', 'B M S', 'B E'],
    'Diploma': ['Diploma', 'Diploma in Electrical Engineering', 'Diploma in Interior Designing', 'Diploma in Mechanical Engineering', 'Diploma in Civil Engineering', 'DCE', 'D Ed', 'BAMS', 'Diploma in Fashion designing', 'Diploma in Instrumentation Engineering'],
    'Legal': ['LLB', 'Master of Law'],
    'PG': ["Post Graduation", 'PG Diploma Marketing management', 'M Sc', 'MBA', 'M E', 'M D', 'PG Diploma Marketing management', 'PG Diploma  Business Admin', 'Ph.D.', 'M A', 'ICWA', 'M S', 'MCM', 'MCA', 'MED', 'M. Phil', 'MMS', 'M. Arch.'],
    'Below 10': ['Illiterate', 'Under Matric (Class l to lX)'],
    'Above 10': ['H S C', 'ITI'],
    'Commerce':['CA','CFA']
}

def map_to_custom_category(value):
    for category, values in custom_categories.items():
        if value in values:
            return category
    return 'Others'
df['Custom_Education_Category'] = df['Application  Life Assured Education'].map(map_to_custom_category)
print("Unique Values in the New Column:", df['Custom_Education_Category'].unique())
```

VIF, or Variance Inflation Factor, is a metric used to assess the severity of multicollinearity in a regression analysis. Multicollinearity occurs when two or more independent variables in a regression model are highly correlated, making it difficult to isolate the individual effect of each variable on the dependent variable.

VIF measures how much the variance of the estimated regression coefficients is increased due to multicollinearity.
A VIF of 1 indicates no multicollinearity. Generally, a VIF greater than 10 is considered a cause for concern, but the threshold can vary.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

X = df[['Applicant\'s Policy Annualized Premium',
       'Auto Debit of Premium Opted Flag(Yes/No)',
       'Application Life Assured BMI',
       'Total Risk Exposure of HDFC Life w.r.t. Life Assured',
       'Application  Life Assured City Tier', 'Application  Life Assured Age',
       'Application  Life Assured Income',
       'Application  Life Assured Marital Status',
       'Application  Life Assured Occupation',
       'Applicant\'s Policy Product Category',
       'Applicant\'s Policy Premium Payment Frequency',
       'Applicant\'s Policy Price Sensitivity(1-\'Yes\', 0-\'No\')',
       'Application Life Assured Residential Status',
       'Applicant\'s Policy Risk Cessation Term',
       'Applicant\'s Policy Rider Opted Flag(1-\'Yes\', 0-\'No\')',
       'Applicant\'s Policy Sum Assured', 'smoker_drinker',
       'Perf+vintage', 'Custom_Education_Category', 'gender_encoded']]

# Create a DataFrame to store the VIF values
vif_data = pd.DataFrame({'Variable': X.columns, 'VIF': [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]})

# Display the result
print(vif_data)
```

Iterateing through the classifiers in dict_classifiers, fits each one on the training data, and records the training score and training time. Finally, it prints the results sorted by the training score in descending order.

```python
from time import perf_counter

no_classifiers = len(dict_classifiers.keys())

def batch_classify(X_train, y_train, verbose=True):
    df_results = pd.DataFrame(data=np.zeros(shape=(no_classifiers, 3)), columns=['classifier', 'train_score', 'training_time'])
    count = 0
    for key, classifier in dict_classifiers.items():
        t_start = perf_counter()
        classifier.fit(X_train, y_train)
        t_end = perf_counter()
        t_diff = t_end - t_start
        train_score = classifier.score(X_train, y_train)
        df_results.loc[count, 'classifier'] = key
        df_results.loc[count, 'train_score'] = train_score
        df_results.loc[count, 'training_time'] = t_diff
        if verbose:
            print("trained {c} in {f:.2f} s".format(c=key, f=t_diff))
        count += 1
    return df_results

df_results = batch_classify(X_train, y_train)
print(df_results.sort_values(by='train_score', ascending=False))
```
